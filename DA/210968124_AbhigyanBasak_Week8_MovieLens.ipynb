### 2. Read the “ratings.csv” file and create a pivot table with index=‘userId’, columns=‘movieId’, values = 'rating'.

```python
#reading the dataset
ratings=pd.read_csv('ratings.csv')
ratings
```

```python
ratings.shape
```

```python
ratings.isnull().sum()
```

```python
#no missing values
```

```python
#pivot table
ratings_pivot=pd.pivot_table(ratings,values='rating',index='userId',columns='movieId')
ratings_pivot
```

```python
#replacing NaN with 0
ratings_pivot.fillna(0,inplace=True)
ratings_pivot
```

### 3. Use cosine similarity for finding similarity among users.

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#converting from dataframe to numpy array
ratings_pivot=ratings_pivot.to_numpy()
ratings_pivot
```

```python
#pairwise distance matrix
pair_dist=pairwise_distances(ratings_pivot, metric='cosine')
pair_dist
```

```python
#converting from numpy array back to dataframe
pair_dist=pd.DataFrame(pair_dist)
pair_dist.head()
```

### 4. Find the 5 most similar user for user with user Id 25.

```python
#top5
pair_dist[25].sort_values(ascending=False).head(5)
```

### 5. Use the “movies” dataset to find out the names of movies, user 1 and user 338 have watched in common and how they have rated each one of them.

```python
movies=pd.read_csv('movies.csv')
movies
```

```python
#merging the ratings and movies dataset
full=pd.merge(ratings,movies,on='movieId')
full
```

```python
#creating dataframes for user1 and user338
user1=full[full.userId==1][['movieId','rating','title']]
user338=full[full.userId==338][['movieId','rating']]
user338.head()
```

```python
user1.shape,user338.shape
```

```python
#merging the two user dataframes on movieID which will result a dataframe of elements with same movieID from the two 
#datasets
common=pd.merge(user1,user338,on='movieId')
common
```

### 7. Create a pivot table for representing the similarity among movies using correlation.

```python
df_movie = full.pivot_table(index='movieId',columns='userId',values='rating')
df_movie.fillna(value=0, inplace=True)
df_movie
```

```python
#Finding the correlation between movies
movie_sim=pd.DataFrame(1-pairwise_distances(df_movie.values,metric="correlation"))
movie_sim.head()
```
