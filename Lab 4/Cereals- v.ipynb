```python
df=pd.read_csv("Cereals.csv")
df
```

```python
df[['protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']].describe()
```

```python
df.mask(df=='?',np.nan)
```

```python
num=pd.to_numeric(df['carbo'],errors='coerce')
df['carbo']=num.fillna(num.mean())
```

```python
num=pd.to_numeric(df['sugars'],errors='coerce')
df['sugars']=num.fillna(num.mean())
```

```python
num=pd.to_numeric(df['potass'],errors='coerce')
df['potass']=num.fillna(num.mean())
```

```python
df
```

```python
nut=df[['protein','fat','sodium','fiber','carbo','sugars','potass','vitamins']]
nut.describe()
```

```python
for col in nut:
    Q1=np.quantile(df[col],0.25)
    Q3=np.quantile(df[col],0.75)
    IQR=Q3-Q1
    lower_bound= Q1-(1.5*IQR)
    upper_bound= Q3+(1.5*IQR)
    Q2= df[col].median()
    df[col]= np.where(df[col]>upper_bound,Q2,df[col])
    df[col]= np.where(df[col]<lower_bound,Q2,df[col])
```
