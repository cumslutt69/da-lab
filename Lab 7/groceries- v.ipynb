Q3- Prepare the data for finding transaction set

```python
df[0] = df[0].apply(lambda x : x.split(","))
df
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

Q5- Find top selling items with minimum support of 2%

```python
te = TransactionEncoder()
te_ary = te.fit(df[0]).transform(df[0])
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
apriori(df, min_support=0.02,use_colnames = True).sort_values(by = "support" , ascending = False).head()
```

Q6-Find all frequent itemsets with minimum support of 5%.

```python
apriori(df, min_support=0.05,use_colnames = True).sort_values(by = "support")
```

Q7- Find all frequent itemsets of length 2 with minimum support of 2%.

```python
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[frequent_itemsets['length']==2]
```

Q8- Find the top 10 association rules with minimum support of 2%, sorted by confidence in descending order

```python
var = apriori(df, min_support=0.02, use_colnames=True)
var = association_rules(var, metric="lift")
var
```

```python
var.sort_values(by="confidence" , ascending =False).head(10)
```

Q9- Find association rules with minimum support of 2% and lift of more than 1.0

```python
var[var['lift']>1]
```
